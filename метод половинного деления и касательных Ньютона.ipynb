```python
# начальные значения

e = 1e-8
a = 3
b = 4
c = -10
```

```python
def f(x):
    return e * x ** 3 + a * x ** 2 + b * x + c
```

```python
def df(x):
    return 3 * e * x ** 2 + 2 * a * x + b
```

```python
def find_extremum_points():
    A = 3 * e
    B = 2 * a
    C = b
    
    discriminant = B**2 - 4 * A * C
    
    if discriminant < 0:
        return None
    elif discriminant == 0:
        x_extremum = -B / (2 * A)
        return [x_extremum]
    else:
        x1 = (-B + (discriminant) ** 0.5) / (2 * A)
        x2 = (-B - (discriminant) ** 0.5) / (2 * A)
        return [x1, x2]
```

```python
def bisection_method(f, a1, a2, tol=1e-8, max_iter=100000):
    
    if f(a1) * f(a2) >= 0:
        raise ValueError("Функция должна иметь разные знаки на концах интервала [a1, a2].")

    iter_count = 0
    while (a2 - a1) / 2 > tol and iter_count < max_iter:
        midpoint = (a1 + a2) / 2
        f_mid = f(midpoint)

        if abs(f_mid) < tol:
            return midpoint

        if f(a1) * f_mid < 0:
            a2 = midpoint
        else:
            a1 = midpoint

        iter_count += 1
    
    return (a1 + a2) / 2
```

```python
def newton_method(f, df, x0, tol=1e-8, max_iter=100000):
    iter_count = 0
    x = x0
    
    while abs(f(x)) > tol and iter_count < max_iter:
        x_new = x - f(x) / df(x)
        
        if abs(x_new - x) < tol:
            return x_new
        
        x = x_new
        iter_count += 1

    return x 
```

```python
x1, x2 = find_extremum_points()
```

```python
a1 = min(x1, x2)
b1 = max(x1, x2)

res1 = bisection_method(f, a1, b1)

print(bisection_method(f, a1, b1), newton_method(f, df, (a1 + b1) / 2))
    
```

```python
a2 = -(10 ** 10)
b2 = a1

res2 = bisection_method(f, a2, b2)

print(bisection_method(f, a2, b2), newton_method(f, df, a2))
```
