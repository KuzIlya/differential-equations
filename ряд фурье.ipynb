```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx, x0, xn, m = 1e-2, -1, 1, 0.33
```

```python
def solution(x: np.ndarray) -> np.ndarray:
    y = np.zeros_like(x)
    for n in range(1, 30):
        a_n = np.trapz(x**2 * np.cos(n * np.pi * x), x)
        y += a_n * np.cos(n * np.pi * x)
    return y
```

```python
plt.plot(np.arange(x0, xn, dx), solution(np.arange(x0, xn, dx)) + m, label='$S_n(x)$')
plt.plot(np.arange(x0, xn, dx), np.arange(x0, xn, dx) ** 2, label='$x^2$')
plt.legend()
plt.show()
```
